Construa um gráfico de linhas com as seguintes características:

- Destaque o impacto da pandemia (2020 e 2021) com um vspan

```python
# listas de anos e refeições
anos = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
refeicoes = [40139, 127020, 168193, 153115, 202102, 232897, 277912, 205350, 233389, 232797, 110023, 160350, 222789]
```

```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

#Paleta de cores
cores = plt.get_cmap('Set3').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle = ciclo_cores)

cor_anotacoes = cores[3]
#Tamanho 
plt.figure(figsize=(14, 6))

#marcadores
#marcadores = np.where(vendas)
for posicao in (0, -1):
    plt.scatter(anos[posicao], refeicoes[posicao], color=cor_anotacoes)
    plt.annotate(refeicoes[posicao], color=cor_anotacoes, xy=(anos[posicao], refeicoes[posicao]), xytext = (8, -3), textcoords = 'offset points')

plt.plot(anos, refeicoes, linewidth = 3)

#Destaque
plt.axvspan(2020, 2021, color = cores[2], alpha = 0.5)
plt.text(2020, 40000, 'Pandemia', fontsize = 11, weight = 'bold', color = cor_anotacoes)

#Titulo
plt.title('Distribuição de refeições por ano')

#configurando anos
plt.xticks(ticks = anos)

plt.box(False)

#Configuração dos ticks
plt.yticks([])
plt.tick_params(axis = 'x', length = 0)

#Salvar
plt.savefig('Grafico_de_linhas.PNG', bbox_inches = 'tight', dpi = 300)

plt.show
```
