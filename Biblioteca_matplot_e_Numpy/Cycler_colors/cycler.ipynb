# Código de ciclo de cores

```python
from cycler import cycler

#paleta de cores
cores = plt.get_cmap('Set1').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle = ciclo_cores)
```

Esse trecho pode ser adicionado em qualquer código que use representações gráficas criadas a partir da biblioteca jupyter/matplotlib...

É usado para rodar a paleta de cores selecionada. Para alterar a paleta é necessário mudar a string na variável cores.
Outras opções de paleta podem ser encontradas em: https://matplotlib.org/stable/users/explain/colors/colormaps.html
