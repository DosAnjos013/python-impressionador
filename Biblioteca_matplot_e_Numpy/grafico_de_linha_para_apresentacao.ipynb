## Gráfico de linha para apresentação


```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

#paleta de cores
cores = plt.get_cmap('Set1').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle = ciclo_cores)

#gráfico
marcadores = np.where(vendas == np.min(vendas), 'x', 'o')

plt.plot(meses, vendas)

for posicao, marcador in enumerate(marcadores):
    if marcador == 'x':
        cor = 'red'
    else:
        cor = 'red'
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, color=cor)

plt.annotate('Menor valor', color = 'red', xy = (2, 78), xytext = (2.2, 79))

#destaque
plt.axvspan(6, 11, color = cores[2], alpha = 0.2)

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
#plt.grid(True)
plt.show()
```
