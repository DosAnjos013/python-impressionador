# Trabalhando com eixos



## Mais de um eixo y em uma figura

Crie uma figura de um gráfico de precipitação com um eixo secundário para a temperatura, para mostrar a relação entre a temperatura e a precipitação em uma cidade.



```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
precipitacao_media = [200, 180, 160, 100, 60, 30, 30, 35, 60, 100, 160, 200]
temperatura_media = [30, 29, 27, 25, 23, 22, 22, 23, 24, 25, 26, 28]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

cor_precipitacao = 'darkblue'
cor_temperatura = 'red'

ax.bar(meses, precipitacao_media, color = cor_precipitacao, label = 'Precipitação média (mm)')
ax.set_ylabel('Precipitação média (mm)', color = cor_precipitacao)
ax.tick_params(axis = 'y', labelcolor = cor_precipitacao)

ax2 = ax.twinx()

ax2.plot(meses, temperatura_media, color = cor_temperatura, linewidth = 2, label = 'Temperatura (C°)')
ax2.set_ylabel('Temperatura (C°)', color = cor_temperatura)
ax2.tick_params(axis = 'y', labelcolor = cor_temperatura)
ax2.set_ylim(20, 32)

ax.set_title('Temperatura e precipitação ao longo do ano')

plt.grid(True)
fig.legend(bbox_to_anchor = (0.75,1), bbox_transform=ax.transAxes)

plt.show
```
