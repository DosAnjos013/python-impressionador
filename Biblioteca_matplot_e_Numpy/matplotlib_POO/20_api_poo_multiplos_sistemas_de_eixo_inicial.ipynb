# Trabalhando com eixos



## Gráfico de linhas

```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

fig, ax = plt.subplots()

ax.plot(meses, vendas)
ax.set_xlabel('Mês')
ax.set_ylabel('Vendas (em milhares de reais)')
ax.set_title('Crescimento das Vendas ao Longo do Ano')

plt.show()
```

## Gráfico de barras

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

fig, ax = plt.subplots()

ax.bar(produtos, quantidades)
ax.set_xlabel('Produto')
ax.set_ylabel('Quantidade de vendas')
ax.set_title('Quantidade de vendas por produto')

plt.show()
```

## Gráfico de pizza

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades_vendidas = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia

fig, ax = plt.subplots()

ax.pie(quantidades_vendidas, labels=categorias, explode=destaque, autopct='%.1f')
ax.set_title('Categorias de produtos vendidos')

plt.show()
```

## Gráfico de barras empilhadas

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

fig, ax = plt.subplots()

ax.bar(anos, vendas_online, label='Online')
ax.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

ax.set_xlabel('Ano')
ax.set_ylabel('Vendas (em milhões de reais)')
ax.set_title('Vendas online vs loja')

ax.set_xticks(anos)

ax.legend()

plt.show()
```

# A divisão dos gráficos é igual a um array numpy


```python
fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize = (12, 12))

# Figura 1

axs[0,0].plot(meses, vendas)
axs[0,0].set_ylabel('Vendas em milhares (R$)')
axs[0,0].set_xlabel('Meses')
axs[0,0].set_title('Vendas ao longo do ano')
axs[0,0].grid(True)

# FIgura 2

axs[0,1].bar(produtos, quantidades)
axs[0,1].bar_label(axs[0,1].bar(produtos, quantidades), labels = quantidades)
axs[0,1].set_title('Quantidade de vendas por produto')
axs[0,1].set_ylabel('Vendas')
axs[0,1].set_xlabel('Meses')

# Figura 3

axs[1,0].pie(quantidades_vendidas, labels = categorias, explode = destaque, autopct='%1.1f%%')
axs[1,0].set_title('Categorias de produtos vendidos')

# Figura 4

barra1 = axs[1,1].bar(anos, vendas_online, label = 'Online')
axs[1,1].bar_label(barra1, label_type = 'center')
barra2 = axs[1,1].bar(anos, vendas_loja, bottom = vendas_online, label = 'Loja')
axs[1,1].bar_label(barra2, label_type = 'center')
axs[1,1].set_xlabel('Anos')
axs[1,1].set_ylabel('Vendas')
axs[1,1].set_title('Vendas online X vendas loja')

axs[1,1].legend()
axs[1,1].set_xticks(anos)

plt.show
```

## Mosaico


```python
mosaico = 'AA;BB;CD'
fig = plt.figure(figsize=(20,20))
axs = fig.subplot_mosaic(mosaico)

# Figura 1

axs['A'].plot(meses, vendas)
axs['A'].set_ylabel('Vendas em milhares (R$)')
axs['A'].set_xlabel('Meses')
axs['A'].set_title('Vendas ao longo do ano')
axs['A'].grid(True)

# FIgura 2

axs['B'].bar(produtos, quantidades)
axs['B'].bar_label(axs['B'].bar(produtos, quantidades), labels = quantidades)
axs['B'].set_title('Quantidade de vendas por produto')
axs['B'].set_ylabel('Vendas')
axs['B'].set_xlabel('Meses')

# Figura 3

axs['C'].pie(quantidades_vendidas, labels = categorias, explode = destaque, autopct='%1.1f%%')
axs['C'].set_title('Categorias de produtos vendidos')

# Figura 4

barra1 = axs['D'].bar(anos, vendas_online, label = 'Online')
axs['D'].bar_label(barra1, label_type = 'center')
barra2 = axs['D'].bar(anos, vendas_loja, bottom = vendas_online, label = 'Loja')
axs['D'].bar_label(barra2, label_type = 'center')
axs['D'].set_xlabel('Anos')
axs['D'].set_ylabel('Vendas')
axs['D'].set_title('Vendas online X vendas loja')

axs['D'].legend()
axs['D'].set_xticks(anos)

plt.show
```
