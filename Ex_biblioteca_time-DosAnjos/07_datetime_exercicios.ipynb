## Oferecendo desconto para cliente com base na última compra

Suponha que você está trabalhando para uma empresa que deseja rastrear a atividade do cliente. Uma métrica que eles estão interessados é o tempo que passou desde a última transação do cliente. Se for muito tempo, eles podem oferecer um desconto para o cliente. Crie um script Python que mostra quanto tempo se passou desde a última compra do cliente. Se faz mais de 30 dias, mostre uma mensagem oferecendo um desconto para o cliente.

```python
from datetime import datetime, timedelta
agora = datetime.now()

dias_sem_comprar = timedelta(days=30)
clientes = {'João': '09/03/2025, 15:30:30', 'Danni':'20/04/2025, 12:15:57', 'Blume':'15/02/2025, 16:14:32', 'Chloe':'15/04/2025, 09:36:57', 'Hanna':'10/01/2025, 10:25:31', 'Neguinha':'20/02/2025, 11:30:20'}
formato = '%d/%m/%Y, %H:%M:%S'
for nome, compra in clientes.items():
    ultima_compra = datetime.strptime(compra, formato)
    if (agora - ultima_compra) > dias_sem_comprar:
        print('Mandar mensagem para: {}'.format(nome))
```

## Data e hora em diferentes fusos horários

Uma empresa tem escritórios em São Paulo, Nova York e Tóquio. Crie um script Python que mostra a data e hora atuais nesses três fusos horários. Exiba, também, se estes escritórios estão abertos ou fechados (9h às 17h).


```python
from datetime import datetime
import pytz

fuso_sp = datetime.now(pytz.timezone('America/Sao_Paulo'))
fuso_ny = datetime.now(pytz.timezone('America/New_York'))
fuso_tk = datetime.now(pytz.timezone('Asia/Tokyo'))

fusos = {'são Paulo':fuso_sp, 'Nova Iorque':fuso_ny, 'Tóquio':fuso_tk}

for cidade, fuso in fusos.items():
    if 17 > fuso.hour >= 9 :
        print(f'O escritório de {cidade} está aberto!\n')
    else:
        print(f'O escritório de {cidade} está fechado!\n')
```
