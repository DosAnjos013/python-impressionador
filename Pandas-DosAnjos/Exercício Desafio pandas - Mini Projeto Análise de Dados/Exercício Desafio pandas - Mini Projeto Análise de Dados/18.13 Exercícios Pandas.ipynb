#### Leitura dos arquivos CSV


```python
import pandas as pd
funcionarios_df = pd.read_csv(r'CadastroFuncionarios.csv', sep = ';', decimal = ',')
clientes_df = pd.read_csv(r'CadastroClientes.csv', sep = ';', decimal = ',')
servicos_df = pd.read_excel('BaseServiçosPrestados.xlsx')
display(servicos_df)
display(clientes_df)
display(funcionarios_df)
```

### 1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa? <br>
    Sugestão: calcule o salário total de cada funcionário, salário + benefícios + impostos, depois some todos os salários

```python
total_salarios_df = funcionarios_df[['Salario Base', 'Impostos', 'Beneficios', 'VT', 'VR']].sum().sum().round(2)
#Bagunça feita para poder trazer o resultado para o padrão brasileiro... sendo o separador de milhar a virgula e o separador de casas decimais o ponto!
print(f'O valor total gasto com a folha salarial é R${total_salarios_df:,.2f}'.replace(',','x').replace('.',',').replace('x','.'))
```

### 2. Qual foi o faturamento da empresa?<br>
    Sugestão: calcule o faturamento total de cada serviço e depois some o faturamento de todos

```python
faturamento_df = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente', 'Valor Contrato Mensal']])
print('O faturamento total foi R${:,.2f}'.format(sum(faturamento_df['Tempo Total de Contrato (Meses)'] * faturamento_df['Valor Contrato Mensal'])).replace(',','x').replace('.',',').replace('x','.'))
```

### 3. Qual o % de funcionários que já fechou algum contrato?<br>
    Sugestão: na base de serviços temos o funcionário que fechou cada serviço. Mas nem todos os funcionários que a empresa tem já fecharam algum serviço.<br>
    . Na base de funcionários temos uma lista com todos os funcionários<br>
    . Queremos calcular Qtde_Funcionarios_Fecharam_Serviço / Qtde_Funcionários_Totais<br>
    . Para calcular a qtde de funcionários que fecharam algum serviço, use a base de serviços e conte quantos funcionários tem ali. Mas lembre-se, cada funcionário só pode ser contado uma única vez.<br><br>
    Dica: se você aplicar o método .unique() em uma variável que é apenas 1 coluna de um dataframe, ele vai excluir todos os valores duplicados daquela coluna.<br>
    Ex: unicos_colunaA = dataframe['colunaA'].unique() te dá como resposta uma lista com todos os itens da colunaA aparecendo uma única vez. Todos os valores repetidos da colunaA são excluidos da variável unicos_colunaA 

```python

def perc_funcionarios_df(df_servicos, df_funcionarios):
    venderam = df_servicos['ID Funcionário'].nunique()
    total = len(df_funcionarios['ID Funcionário'])
    percentual = (venderam/total)
    return percentual
print('{:.2%} dos funcionários ja efetivou pelo menos 1 venda'.format(perc_funcionarios_df(servicos_df, funcionarios_df)))
```

### 4. Calcule o total de contratos que cada área da empresa já fechou

```python
def contratos_por_area():
    combinado_df = funcionarios_df[['ID Funcionário','Nome Completo', 'Area']].merge(servicos_df[['ID Funcionário']])
    return(combinado_df.value_counts('Area'))

print(contratos_por_area())
contratos_por_area().plot(kind='bar', rot=45, title='Total de contratos fechados por área', xlabel='Áreas', ylabel='Quantidade de contratos')
```

### 5. Calcule o total de funcionários por área

```python
print(funcionarios_df.value_counts('Area'))
funcionarios_df.value_counts('Area').plot(kind='bar',rot=45, ylabel='Quantidade funcionários', xlabel='Áreas', title='Quantidades de funcionários por área')
```

### 6. Qual o ticket médio mensal (faturamento médio mensal) dos contratos?

```python
def ticket_medio(valores_de_contrato):
    tm = ('O ticket médio mensal dos contratos é: R${:,.2f}'.format(valores_de_contrato['Valor Contrato Mensal'].mean().round(2)).replace(',','x').replace('.',',').replace('x','.'))
    return (tm)
print(ticket_medio(clientes_df))
```
