### Exercício 1
Crie um sistema de consulta de preços
Seu sistema deve:
- Pedir para o usuário o nome de um produto
- Caso o produto exista na lista de produtos, o programa deve retornar o preço do produto como resposta
       - Ex: O produto celular custa R$1500
- Caso o produto não exista na lista de produtos, o programa deve printar uma mensagem para o usuário tentar novamente


```python
precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 2000}
produtos = precos.keys()
print ('Produtos disponíveis: {}'.format(", ".join(list(produtos))))

consulta = True
while consulta == True:
    produto = input('\nDigite o nome do produto que deseja consultar: ')
    produto = produto.casefold()
    produto = produto.strip()
    if produto in produtos:
        print ('*' * 40)
        print('O preço do {} é R$: {}'.format(produto, precos[produto]))
        print ('*' * 40)
        continua = input('Deseja consultar outro produto? (1/sim)|(2/não): ')
        if continua == '1' or continua == sim:
            continue
        else: consulta = False
    else:
        print('Produto não encontrado!')
        continua = input('Deseja tentar novamente? (1/sim)|(2/não): ')
        if continua == '1' or continua == sim:
            continue
        else: consulta = False
```
