Crie uma function que calcule qual o % de carga tributária que está sendo aplicado sobre um determinado produto, dado o preço de venda, o "lucro" e os custos (com exceção do imposto) dele.

```python
import time
'''Para realizar esse cálculo a conta é:

imposto = preço de venda - (custo + lucro)

Porcentagem(%) =  (imposto/Preço de venda) * 100

'''
inicio = time.time()
def calcular_imposto(valor_venda, valor_compra, valor_lucro):
    #alternativa para facilitar o entendimento porém, após alguns testes, fica claro que o modo de execução atual é mais rápido.
    '''
    valor_imposto = valor_venda - (valor_compra + valor_lucro)
    porc_imposto = (valor_imposto / valor_venda) * 100
    return (porc_imposto)
    '''
    
    return(((valor_venda - (valor_compra + valor_lucro)) / valor_venda))# <- João se você multiplicar por 100 aqui, sua formatação tem que ser {:.2f}, mas se você não multiplicar, sua formataçõa deve ser {:.2%}


preco = 1500
custo = 400
lucro = 800

#alternativa para facilitar o entendimento porém, após alguns testes, fica claro que o modo de execução atual é mais rápido.

#imposto = calcular_imposto(preco, custo, lucro)
#print('O imposto do produto é {:,.2f}%'.format(imposto))


print('O imposto do produto é {:.2%}'.format(calcular_imposto(preco, custo, lucro)))
fim = time.time()
print('O tempo necessário para executar o código foi {:.10f}'.format(fim-inicio))
```
