# Principal Aplicação de Lambda Expressions

### Usar lambda como argumento de alguma outra função, como map e filter

```python
preco_tecnologia = {'notebook asus': 2450, 'iphone': 4500, 'samsung galaxy': 3000, 'tv samsung': 1000, 'ps5': 3000, 'tablet': 1000, 'notebook dell': 3000, 'ipad': 3000, 'tv philco': 800, 'notebook hp': 1700}
```

### map()
- Queremos saber o preço de cada produto adicionando o valor do imposto de 30% sobre o valor do produto

```python
#Fazendo por function
def calcula_imposto(dicionario):
    preco_tecnologia_atualizado = {}
    for chave, valor in dicionario.items():
        preco_tecnologia_atualizado[chave] = valor * 1.30
    return preco_tecnologia_atualizado

print(calcula_imposto(preco_tecnologia))
```

```python
#fazendo por lambda
calc_imposto = dict(map(lambda item: (item[0], item[1] * 1.30), preco_tecnologia.items()))
print(calc_imposto)
```

### filter()
- Queremos apenas os produtos que custam acima de 2000

```python
#fazendo por function
def acima_valor(dicionario):
    acima_2000 = {}
    for chave, valor in dicionario.items():
        if valor > 2000:
            acima_2000[chave] = valor
        else: 
            pass
    return acima_2000

print(acima_valor(preco_tecnologia))
```

```python
#fazendo por lambda
menor = dict(filter(lambda x : x[1] > 2000, preco_tecnologia.items()))
print(menor)
```
