```python
rng = np.random.default_rng(seed = 42)
funcionarios = ["João", "Washington", "Danielle", "Elizangela", "Chloe", "Blume", "Hanna", "Neguinha", "Norminha", "Scobby"]
array_funcionarios = np.array(funcionarios)
salarios = rng.integers(low = 1500, high = 5000, size = 10)
print(salarios)
```

```python
media_salarios = np.mean(salarios)
print(media_salarios)
```

```python
salarios_acima_media = np.where(salarios > media_salarios)
print(salarios_acima_media)
```

```python
print(", ".join(array_funcionarios[salarios_acima_media]),"tem salarios acima da média!")
```
